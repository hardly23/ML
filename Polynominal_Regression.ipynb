```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
plt.style.use("dark_background")
```

```python
posi=pd.read_csv("Position_Salaries.csv")
x=posi.iloc[ :,1:-1].values
y=posi.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression
lin_1 =LinearRegression()
lin_1.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
pol_1 =PolynomialFeatures(degree=5)
x_pol=pol_1.fit_transform(x)
lin_2=LinearRegression()
lin_2.fit(x_pol,y)
```

```python
plt.scatter(x,y,color="blue")
plt.plot(x,lin_1.predict(x),color="red")
```

```python
plt.scatter(x,y,color="blue",edgecolors=None)
plt.plot(x,lin_2.predict(pol_1.fit_transform(x)),color="red")
```
